# Data Visulization 


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Fare Trends Over the Years
plt.figure(figsize=(10, 6))
fare_trends = df_cleaned.groupby('Year')['fare'].mean()
sns.lineplot(x=fare_trends.index, y=fare_trends.values, marker='o', color='b')
plt.title("Average Fare Trends Over the Years (1993-2024)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Fare (USD)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Passenger Distribution by City (Top 10 Cities)
plt.figure(figsize=(10, 6))
top_cities = df_cleaned.groupby('city1')['passengers'].sum().nlargest(10)
sns.barplot(x=top_cities.values, y=top_cities.index, palette='viridis')
plt.title("Top 10 Cities by Passenger Count", fontsize=14)
plt.xlabel("Passenger Count", fontsize=12)
plt.ylabel("City", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
carrier_comparison = df_cleaned[['fare_lg', 'fare_low']].mean()

# 3. Comparison of Large vs Low-Fare Airlines
sns.barplot(x=carrier_comparison.index, y=carrier_comparison.values, hue=carrier_comparison.index, palette='coolwarm', dodge=False, legend=False)
plt.title("Average Fares: Large vs Low-Fare Airlines", fontsize=14)  # Title with fontsize
plt.xlabel("Carrier Type", fontsize=12)
plt.ylabel("Average Fare (USD)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Calculate total passengers for large and low-fare airlines by year
market_share_df = df_cleaned.groupby('Year').agg({
    'passengers': 'sum', 
    'large_ms': 'mean', 
    'lf_ms': 'mean'
}).reset_index()

# Calculate market share as a percentage for large and low-fare airlines
market_share_df['large_ms_share'] = market_share_df['large_ms'] * 100
market_share_df['lf_ms_share'] = market_share_df['lf_ms'] * 100

# Plot the market share comparison
plt.figure(figsize=(10, 6))
sns.lineplot(x=market_share_df['Year'], y=market_share_df['large_ms_share'], label='Large Airlines', marker='o', color='b')
sns.lineplot(x=market_share_df['Year'], y=market_share_df['lf_ms_share'], label='Low-Fare Airlines', marker='o', color='r')

plt.title('Market Share: Large vs Low-Fare Airlines (1993-2024)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Market Share (%)', fontsize=12)
plt.legend(title="Airline Type")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
